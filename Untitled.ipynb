```python
import os
from langchain.tools import Tool
from langchain.prompts import PromptTemplate
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_groq import ChatGroq
from langchain.agents import AgentExecutor, create_react_agent, initialize_agent, AgentType
import requests

# Load API Keys
os.environ["GROQ_API_KEY"] = "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"

# Load Groq LLM
llm = ChatGroq(model="llama3-8b-8192", temperature=0.7)

# Define the custom HR Trend Research tool
class HRTrendsResearchTool(Tool):
    def _run(self, *args, **kwargs):
        return self.research_hr_trends()

    def _arun(self, *args, **kwargs):
        raise NotImplementedError("Asynchronous execution not supported for this tool.")
    
    def research_hr_trends(self):
        """Fetches trending HR topics for the latest year using Serper API."""
        try:
            url = "https://google.serper.dev/search"
            headers = {"X-API-KEY": "77730b7a48074abb7315812bc481e49366418abc", "Content-Type": "application/json"}
            params = {"q": "Latest trending HR topics 2025", "num": 5}

            response = requests.post(url, json=params, headers=headers)
            data = response.json()

            if "organic" in data:
                return [result["title"] for result in data["organic"]]  # Extract titles
            else:
                return ["No relevant HR trends found."]
        
        except Exception as e:
            return [f"Error fetching HR trends: {e}"]

# Content Planning Agent: Creates a structured outline based on trends
def generate_outline(topic):
    """Creates a detailed blog outline based on HR trends."""
    prompt = f"Create a detailed, structured blog outline for the topic 'HR trends in 2025':\n\n{topic}"
    return llm.invoke(prompt).content  # Extract the text response

# Content Generation Agent: Writes the full blog
def generate_blog(outline):
    """Generates a full blog based on the outline."""
    prompt = f"Write a detailed blog based on this outline:\n\n{outline}"
    return llm.invoke(prompt).content  # Extract the text response

# SEO Optimization Agent: Ensures the blog follows SEO best practices
def seo_optimization(content):
    """Optimize the blog content for SEO best practices."""
    prompt = f"Optimize the following blog content for SEO:\n\n{content}"
    return llm.invoke(prompt).content  # Extract the text response

# Review Agent: Proofreads and enhances the blog content
def review_content(content):
    """Proofreads and enhances the blog content for readability and quality."""
    prompt = f"Proofread and enhance the following blog content to make it clear, engaging, and suitable for publication:\n\n{content}"
    return llm.invoke(prompt).content  # Extract the text response

# Define the tools for the multi-agent system
tools = [
    HRTrendsResearchTool(
        name="HR Trends Research", 
        func=HRTrendsResearchTool.research_hr_trends,
        description="Fetches the latest trending HR topics for 2025."
    ),
    Tool(
        name="Content Planning", 
        func=generate_outline,
        description="Creates a detailed blog outline based on HR trends."
    ),
    Tool(
        name="Content Generation", 
        func=generate_blog,
        description="Generates a full blog based on the outline."
    ),
    Tool(
        name="SEO Optimization", 
        func=seo_optimization,
        description="Optimizes the blog content for SEO best practices."
    ),
    Tool(
        name="Review", 
        func=review_content,
        description="Proofreads and enhances the blog content for readability."
    )
]

# Initialize the agent executor with the tools
agent = initialize_agent(
    tools=tools,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    llm=llm,
    verbose=True
)

# Run the agent (this will invoke the entire pipeline)
def run_blog_pipeline():
    print("🌟 Starting the multi-agent workflow...\n")
    # Fetch HR trends
    hr_trends = agent.run("Fetch trending HR topics for 2025.")
    print(f"HR Trends Found: {hr_trends}\n")
    
    # Generate the blog outline based on trends
    outline = generate_outline(hr_trends)
    print(f"Generated Outline: {outline}\n")
    
    # Generate the blog content
    blog_content = generate_blog(outline)
    print(f"Generated Blog Content: {blog_content}\n")
    
    # Optimize for SEO
    seo_optimized_content = seo_optimization(blog_content)
    print(f"SEO Optimized Content: {seo_optimized_content}\n")
    
    # Review the content
    reviewed_content = review_content(seo_optimized_content)
    print(f"Reviewed Content: {reviewed_content}\n")
    
    print("\n✅ Final Blog Generated Successfully!\n")
    return reviewed_content

# Run the full blog pipeline
final_blog_output = run_blog_pipeline()
print("\n🌟 Final Blog Output:\n", final_blog_output)
```
